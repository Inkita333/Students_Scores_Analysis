```python
plt.figure(figsize=(5, 5))
ab=sns.countplot(data=df, x='Gender')
ab.bar_label(ab.containers[0])
plt.title('Gender Distribution')
plt.show()
```

### From the above graph we have analyzed that the numbers of females in the data is more than the number of males

```python
gb=df.groupby('ParentEduc').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
```

```python
print(gb)
```

```python
sns.heatmap(gb, annot=True)
plt.title("Relationship Between Parent's Education and Student's Score")
plt.show()
```

### from the above chart we have concluded that the education of a parent have a good impact on their Scores

```python
gb1=df.groupby('ParentMaritalStatus').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
gb1
```

```python
sns.heatmap(gb1, annot=True)
plt.title("Relationship Between Parent's Maritual Status and Student's Score")
plt.show()
```

### From the above chart we have concluded that their is no/negligible impact on the student's scores due to their parent's marital status

### Ethnic Group Distribution

```python
groupA=df.loc[(df['EthnicGroup']=='group A')].count()
groupB=df.loc[(df['EthnicGroup']=='group B')].count()
groupC=df.loc[(df['EthnicGroup']=='group C')].count()
groupD=df.loc[(df['EthnicGroup']=='group D')].count()
groupE=df.loc[(df['EthnicGroup']=='group E')].count()
mlist=[groupA['EthnicGroup'], groupB['EthnicGroup'], groupC['EthnicGroup'], groupD['EthnicGroup'], groupE['EthnicGroup']]
l=['groupA', 'groupB', 'groupC', 'groupD', 'groupE']
plt.pie(mlist, labels=l, autopct='%1.2f%%')
plt.title('Distribution of Ethenic Groups')
plt.show()
```

```python
gb2=df.groupby('EthnicGroup').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
gb2
```

### Relationship Between Ethenic Group

```python
sns.heatmap(gb2, annot=True)
plt.show()
```

### From the above chart we have concluded that group E has the highest scores and group A has lowest scores in all three subjects compared to other groups

```python
gb3=df.groupby('LunchType').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
gb3
```

```python
sns.heatmap(gb3, annot=True)
plt.title('Relationship between Lunch type and students scores')
plt.show()
```
